```python
from desc import set_device
# set_device("gpu")
from desc.coils import CoilSet
from desc.plotting import plot_coils
```

```python
coils = CoilSet.from_makegrid_coilfile("coils.biot_savart_opt_nfp3.txt",check_intersection=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
for coil in coils:
    data = coil.compute("x",basis="rpz")
    print(np.max(data["x"][:,1]))
```

```python
plt.figure()
for coil in coils:
    data = coil.compute("x",basis="xyz")
    plt.plot(data["x"][:,0], data["x"][:,1])
    plt.xlabel("x")
    plt.ylabel("y")
    
```

```python
coils_with_symmetry = CoilSet(coils[0:3], sym=True, NFP=3)
```

```python
plt.figure()
for coil in CoilSet.from_symmetry(coils_with_symmetry,NFP=3,sym=True,check_intersection=False):
    data = coil.compute("x",basis="xyz")
    plt.plot(data["x"][:,0], data["x"][:,1])
    plt.xlabel("x")
    plt.ylabel("y")
    
```
